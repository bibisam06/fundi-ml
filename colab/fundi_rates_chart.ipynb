```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("funding_rates.csv")
```

```python
print(df['funding_time'])
```

```python
plt.plot(df['funding_time'], df['funding_rate'])
plt.title("BTC Funding Rate(BINANCE)")
plt.xlabel("Time")
plt.ylabel("Funding Rate")
```

```python
plt.show()
```

```python
df.describe()

# 평균 펀딩비 최댓값, 최솟값 등을 출력해주는 함수
```

```python
df_unique = df.drop_duplicates(subset="funding_time")
print(df_unique)

# df_unique => dataset 에서 중복 제거
```

```python
upper = df_unique['funding_rate'].quantile(0.95)
lower = df_unique['funding_rate'].quantile(0.05)

extreme = df_unique[(df_unique['funding_rate'] > upper) | (df_unique['funding_rate'] < lower)]

# 극단값 출력하기
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_unique['funding_time'], df_unique['funding_rate'])
plt.scatter(extreme['funding_time'], extreme['funding_rate'], color='red')
plt.title("Extreme Funding Rates")
plt.show()
```

```python
df['funding_change'] = df['funding_rate'].diff()
plt.plot(df["funding_time"], df["funding_change"])
```

```python
# 이동평균선 ( 3 ) 
df['funding_ma3'] = df['funding_rate'].rolling(3).mean()
```
